# markers

```python
import numpy as np
import matplotlib.pyplot as p

#if u want to mark or point(10,20) and (30,40) the arrays u should be passing should be are

x=np.array([10,30]) #one array with  x coordinates and other with the y coordinate
y=np.array([20,40])
```

```python
p.plot(x,y,marker="h")#marker is like marks points
```

line color

```python
p.plot(x,y,color="pink") #or use just c instead of color
```

```python
p.plot(x,y,marker="o",markeredgecolor="pink")#or use mec instead of markeredgecolor
```

```python
p.plot(x,y,marker="o",markerfacecolor="red") #use mfc instead of markerfacecolor
```

```python
p.plot(x,y,marker="o",markersize="50") #or use ms instead of markersize
```
